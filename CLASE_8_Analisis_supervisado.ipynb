# 1. Modelo analítico de Ciencia de Datos

## 1.1 Definición de modelo analítico

Un **modelo analítico** en el contexto de Ciencia de Datos es:

 Una representación matemática o computacional que describe la relación entre variables de un problema real, con el objetivo de predecir valores, clasificar casos o descubrir patrones sobre datos históricos.  

 - En Ciencia de Datos, el modelo analítico se construye a partir de datos (entradas X y salidas y) mediante técnicas estadísticas y de Machine Learning.  
 - El resultado es un artefacto (función o algoritmo entrenado) que, dada nueva información, permite generar una predicción o clasificar nuevos casos automáticamente.

**Ejemplos de modelos analíticos**:  
- Un modelo de regresión lineal que predice el precio de una vivienda en función de su superficie y ubicación.  
- Un clasificador basado en árboles de decisión que determina si un correo es spam o no spam.  
- Un algoritmo de clustering que agrupa clientes según su comportamiento de compra.


## 1.2 Etapas de un proyecto de ciencia de datos

A continuación, se presentan las etapas clásicas de un proyecto de Ciencia de Datos, desde la definición del problema hasta el mantenimiento del modelo en producción:

1. **Definición del problema y objetivos**  
   - _¿Qué se quiere resolver?_  
   - Ejemplos:  
     - Predecir la demanda de un producto.  
     - Detectar fraudes en transacciones financieras.  
     - Segmentar clientes según su comportamiento.

2. **Recolección y adquisición de datos**  
   - Fuentes de datos:  
     - Bases de datos internas (SQL, NoSQL).  
     - Archivos planos (CSV, Excel).  
     - APIs externas (por ejemplo, API de Twitter).  
     - Web scraping (datos de páginas web).

3. **Procesamiento y limpieza (Data Wrangling)**  
   - Tratamiento de valores faltantes o nulos (imputación, eliminación).  
   - Eliminación de duplicados y corrección de errores de formato.  
   - Transformaciones necesarias:  
     - Conversión de tipos de datos.  
     - Normalización o estandarización.  
     - Codificación de variables categóricas (one-hot encoding, label encoding).

4. **Análisis Exploratorio de Datos (EDA)**  
   - Estadísticas descriptivas (media, mediana, desviación estándar).  
   - Visualizaciones:  
     - Histogramas, boxplots, diagramas de dispersión.  
     - Matriz de correlaciones.  
   - Detección de outliers y patrones iniciales.

5. **Selección y diseño del modelo analítico**  
   - Elección del tipo de modelo (regresión, clasificación, clustering, series de tiempo, etc.).  
   - Definir las variables predictoras (features) y la variable objetivo (target).  
   - Considerar la complejidad del algoritmo según el volumen de datos y dimensionalidad.

6. **Entrenamiento y validación**  
   - División de datos en conjuntos de entrenamiento y prueba (train/test split).  
   - Opcional: validación cruzada k-fold para estimar robustez.  
   - Ajuste de hiperparámetros (GridSearchCV, RandomizedSearchCV).

7. **Evaluación del modelo**  
   - Métricas de regresión:  
     - MAE (Error Absoluto Medio), MSE (Error Cuadrático Medio), R² (Coeficiente de determinación).  
   - Métricas de clasificación:  
     - Accuracy, Precision, Recall, F1‐score, ROC‐AUC.  
   - Análisis de curva de aprendizaje para diagnosticar overfitting/underfitting.

8. **Implementación / Deploy**  
   - Convertir el modelo entrenado en un servicio:  
     - API REST (por ejemplo, Flask, FastAPI).  
     - Integración con aplicaciones web o sistemas internos.  
   - Documentación del modelo y entrega de un manual de uso.

9. **Mantenimiento y monitoreo**  
   - Monitorizar el rendimiento en producción (drift de datos, degradación de métricas).  
   - Reentrenar periódicamente si cambian las condiciones (nuevos datos).  
   - Asegurar reproducibilidad (versionar datos y código).


# 2 Introducción a Machine Learning

## 2.1 ¿Qué es Machine Learning?



**Machine Learning (ML)** es el conjunto de técnicas que permiten a un sistema:
- “Aprender” patrones a partir de datos históricos.
- Generalizar dicho aprendizaje para predecir o tomar decisiones sobre datos nuevos, sin estar programado explícitamente para cada caso.

En lugar de escribir reglas fijas, en ML se expone al algoritmo a ejemplos (datos) y éste ajusta sus parámetros internamente para minimizar un criterio de error. Así obtiene un modelo que, cuando recibe nuevos datos, produce predicciones o clasificaciones.

**Diferencia con la estadística tradicional**:
- La estadística descriptiva/ inferencial pone énfasis en probar hipótesis y describir distribuciones subyacentes.
- ML se enfoca en construir modelos predictivos que “performen” bien al generalizar a datos no vistos, a menudo con datasets de mayor tamaño y complejidad.


## 2.2 Elementos clave de un problema de ML



1. **Datos (X, y)**
   - `X`: matriz de características (features) o variables independientes.  
   - `y`: vector de etiquetas (target) o variables dependientes que queremos predecir.

2. **Modelo (h)**
   - Función matemática parametrizada (por ejemplo, regresión lineal, árbol de decisión) que mapea X → ŷ.  
   - Se denota a menudo como ŷ = h(X; θ), donde θ son los parámetros a ajustar.

3. **Función de pérdida / costo**
   - Mide el error entre predicciones ŷ y valores reales y.  
   - Ejemplos:
     - **Regresión**: MSE (Mean Squared Error), MAE (Mean Absolute Error).  
     - **Clasificación**: cross‐entropy (log loss), hinge loss.

4. **Algoritmo de optimización**
   - Método para ajustar θ y minimizar la función de pérdida.  
   - Ejemplos: descenso de gradiente, gradiente estocástico (SGD), Adam, etc.

5. **Métrica de evaluación**
   - Valor que cuantifica desempeño del modelo sobre datos de validación/prueba.  
   - Ejemplos:
     - **Clasificación**: accuracy, precision, recall, F1‐score, ROC‐AUC.  
     - **Regresión**: R², RMSE, MAE.


## 2.3. Flujo general de un proyecto de ML



A grandes rasgos, un pipeline de ML sigue estos pasos:

1. **Recolección y preparación de datos**  
   - Recolectar datos crudos.  
   - Limpiar valores faltantes, duplicados y errores.  
   - Transformar o codificar variables (one‐hot, label encoding, escalado).

2. **Definir X e y**  
   - Seleccionar las columnas que se usarán como features (`X`).  
   - Definir la columna objetivo a predecir (`y`).

3. **Dividir los datos**  
   - Separar en **entrenamiento** y **prueba** (por ejemplo, 70 % / 30 %).  
   - Opcional: validación cruzada (k‐fold CV) para estimar robustez de métricas.

4. **Entrenamiento (fit)**  
   - Ajustar el modelo a los datos de entrenamiento.  
   - El modelo minimiza la función de pérdida según el algoritmo de optimización.

5. **Predicción (predict)**  
   - Generar predicciones (`ŷ`) sobre datos de prueba (o validación).  

6. **Evaluación**  
   - Calcular métrica(s) de desempeño en el conjunto de prueba.  
   - Analizar si el modelo generaliza bien (evitar overfitting/underfitting).

7. **Ajuste de hiperparámetros (tuning)**  
   - Probar diferentes combinaciones de hiperparámetros (GridSearchCV, RandomizedSearchCV).  
   - Seleccionar la configuración que optimice la métrica deseada.

8. **Validación final y retraining**  
   - Reentrenar con el conjunto completo (train + validación) usando los mejores hiperparámetros.  
   - Evaluar una última vez sobre el conjunto de prueba definitivo.

9. **Deploy / Implementación**  
   - Convertir el modelo entrenado en un servicio (API REST, microservicio).  
   - Documentar su uso y versionar el código y datos.

10. **Monitoreo y mantenimiento**  
    - Vigilar desempeño en producción (drift de datos, degradación de métrica).  
    - Reentrenar periódicamente o cuando cambien las condiciones.


# 3 Tipos de aprendizaje en Machine Learning

## 3.1 Aprendizaje supervisado

**Definición:**
- El aprendizaje supervisado consiste en entrenar un modelo usando entradas (features) X y salidas (etiquetas) y conocidas.
- El algoritmo “aprende” la relación X→y a partir de ejemplos etiquetados.
- Se utiliza para predicción de variables continuas (regresión) o categóricas (clasificación).

**Características clave:**
1. **Datos etiquetados**: cada muestra de entrenamiento incluye la etiqueta asociada.
2. **Objetivo**: aproximar la función f: X → y que minimice el error en datos no vistos.
3. **Entrenamiento**: el modelo ajusta parámetros para disminuir la función de pérdida.
4. **Evaluación**: se usan métricas como accuracy, precision, recall, F1 (clasificación) o MSE, R² (regresión).

**Ejemplos de algoritmos (supervisados):**
- Regresión: Regresión Lineal, Regresión Polinómica, SVR, Random Forest Regressor.
- Clasificación: Regresión Logística, K-Nearest Neighbors, SVM, Árboles de Decisión, Random Forest, Gradient Boosting.


## 3.2 Aprendizaje no supervisado

**Definición:**
- En el aprendizaje no supervisado, solo se tiene un conjunto de entradas X sin etiquetas y.
- El objetivo es descubrir estructura, patrones o agrupaciones dentro de los datos sin guía externa.

**Características clave:**
1. **Datos no etiquetados**: solo disponemos de variables independientes.
2. **Objetivos frecuentes**:
   - **Clustering**: agrupar observaciones en grupos (clusters) similares. Ej.: K-Means, DBSCAN, Hierarchical Clustering.
   - **Reducción de dimensionalidad**: proyectar datos a menos dimensiones manteniendo la mayor varianza posible. Ej.: PCA, t-SNE.
   - **Detección de anomalías**: identificar puntos que difieren significativamente del resto.

**Ejemplo de uso:**
- Segmentación de clientes sin información previa de grupos.
- Análisis de temas (topic modeling) en documentos (LDA).
- Reducción de dimensionalidad para visualización o preprocesamiento.


````python
# Ejemplo práctico: Clustering con K-Means en el dataset Iris (sin usar la etiqueta)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# 1) Cargar Iris
iris = load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)

# 2) Seleccionar solo dos features para visualización (petal length y petal width)
X_cluster = df_iris[['petal length (cm)', 'petal width (cm)']]

# 3) Instanciar K-Means con k=3 (buscamos 3 clusters)
kmeans = KMeans(n_clusters=3, random_state=42)

# 4) Ajustar el modelo
kmeans.fit(X_cluster)

# 5) Obtener etiquetas de cluster
clusters = kmeans.labels_

# 6) Agregar las etiquetas al DataFrame para graficar
df_iris['cluster'] = clusters

# 7) Visualizar los clusters
plt.figure(figsize=(6,4))
sns.scatterplot(
    data=df_iris,
    x='petal length (cm)',
    y='petal width (cm)',
    hue='cluster',
    palette='deep'
)
plt.title('K-Means (k=3) sobre Iris (Petal features)')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Cluster')
plt.show()
````

##3.3 Aprendizaje semisupervisado

**Definición:**
- Combina datos etiquetados y no etiquetados durante el entrenamiento.
- Útil cuando etiquetar es costoso o lleva mucho tiempo, pero se dispone de muchos datos “sin etiqueta”.

**Enfoques comunes:**
1. **Self-training**:  
   - Se entrena un modelo inicial con las pocas muestras etiquetadas, luego se predice en el conjunto no etiquetado.
   - Se agregan las predicciones con alta confianza al conjunto etiquetado, iterando.

2. **Co-training**:  
   - Se trabajan dos modelos diferentes con vistas distintas de los datos.
   - Cada modelo etiqueta ejemplos no etiquetados para entrenar al otro.

3. **Modelos generativos**:  
   - Se usa la distribución conjunta P(X, y) estimada con modelos probabilísticos.

**Ejemplo de aplicación:**
- Clasificación de correos electrónicos donde solo se han etiquetado manualmente unos pocos, pero hay miles de correos sin etiquetar.


## 3.4 Aprendizaje por refuerzo

**Definición:**
- Un agente aprende a tomar decisiones interactuando con un entorno.
- No hay etiquetas directas “X→y” de antemano; en cambio, el agente recibe recompensas o penalizaciones según sus acciones.
- El objetivo es maximizar la recompensa acumulada a lo largo del tiempo.

**Componentes clave:**
1. **Agente**: entidad que toma acciones.  
2. **Entorno**: escenario donde actúa el agente.  
3. **Estado (s)**: representación actual del entorno.  
4. **Acción (a)**: elección del agente en ese estado.  
5. **Recompensa (r)**: señal numérica que indica cuán buena es la acción.  
6. **Política (π)**: función que determina la acción a tomar en cada estado.  
7. **Función de valor (V(s))**: valor esperado de recompensa desde un estado.  
8. **Función Q (Q(s, a))**: valor esperado de recompensa al tomar acción a en estado s.

**Algoritmos populares:**
- Q-Learning  
- Deep Q-Networks (DQN)  
- SARSA  
- Policy Gradients (REINFORCE)  
- Actor-Critic (A2C, PPO)

**Ejemplos de aplicación:**
- Videojuegos (ej.: Atari, AlphaGo).  
- Robótica (control de robots, navegación).  
- Sistemas de recomendación que exploran/explotan.

*Nota:* No incluimos código aquí, pero reconocer que RL no utiliza etiquetas fijas: el agente aprende a base de prueba/error.


| Tipo               | Datos requeridos                | Objetivo principal                       | Ejemplos de algoritmos                      |
|--------------------|---------------------------------|------------------------------------------|---------------------------------------------|
| Supervisado        | X (features), y (etiquetas)     | Predecir o clasificar                    | Regresión Lineal, Regresión Logística, KNN, SVM, Árboles, Random Forest                      |
| No Supervisado     | X (features)                    | Descubrir patrones/agrupaciones/anomalías| K-Means, DBSCAN, PCA, t-SNE                 |
| Semi-Supervisado   | Pocas etiquetas + muchos no etiquetados | Aprovechar data no etiquetada para mejorar el modelo | Self-Training, Co-Training, Modelos generativos |
| Refuerzo (RL)      | Interacción agente-entorno, recompensas | Aprender política para maximizar recompensa | Q-Learning, DQN, SARSA, PPO, Actor-Critic   |


# 4 Aprendizaje supervisado

## 4.1 Flujo de trabajo


En Aprendizaje Supervisado, el flujo completo abarca:

1. **Selección de variables (Feature Engineering)**  
   - Elegir cuáles columnas serán las “features” (X).  
   - Crear variables derivadas si es necesario (por ejemplo, combinaciones, transformaciones logarítmicas).  
   - Codificar variables categóricas:  
     - **Label Encoding** (asignar 0,1,2… a cada categoría).  
     - **One-Hot Encoding** (columnas binarias por cada categoría).

2. **División del dataset**  
   - Separar en **entrenamiento** y **prueba** (ej.: 70 % – 30 %).  
   - Usar `stratify=y` si es clasificación para conservar proporciones de cada clase.  
   - Opcional: reservar un conjunto de validación o usar validación cruzada (k-fold).

3. **Escalado / Normalización**  
   - En modelos basados en distancia (KNN, SVM) o en regularización (Regresión lineal con L1/L2), conviene estandarizar cada feature:  
     - `StandardScaler()` convierte media=0 y varianza=1.  
     - `MinMaxScaler()` escala a rango [0,1].  

4. **Entrenamiento del modelo**  
   - Instanciar el algoritmo elegido (por ejemplo, `LogisticRegression()`, `RandomForestClassifier()`, `SVR()`, etc.).  
   - Ajustar el modelo con `model.fit(X_train, y_train)`.

5. **Predicción**  
   - Generar predicciones sobre el conjunto de prueba con `model.predict(X_test)` (etiquetas).  
   - En clasificación, opcionalmente usar `model.predict_proba(X_test)` para probabilidades.

6. **Evaluación**  
   - **Clasificación**: accuracy, precision, recall, F1, matriz de confusión, ROC‐AUC.  
   - **Regresión**: MAE (Error Absoluto Medio), MSE/RMSE (Error Cuadrático Medio), R².  
   - Graficar curvas de aprendizaje (errores vs tamaño de train) para detectar overfitting/underfitting.

7. **Ajuste de hiperparámetros (tuning)**  
   - Métodos clásicos: `GridSearchCV`, `RandomizedSearchCV` en scikit‐learn.  
   - Definir la grilla de hiperparámetros (p. ej., `n_neighbors` en KNN; `C` y `kernel` en SVM; `n_estimators`, `max_depth` en Random Forest).

8. **Validación Cruzada**  
   - k‐Fold CV divide el train en k subconjuntos, entrena k veces y promedia métricas.  
   - Usa `cross_val_score` o `cross_validate` en scikit‐learn.




## 4.2. Overfitting (Sobreajuste) vs Underfitting (Subajuste)

1. **Underfitting**  
   - Ocurre cuando el modelo es demasiado simple para capturar la estructura subyacente de los datos.  
   - Alto **sesgo**: el modelo no ajusta ni los datos de entrenamiento ni los de prueba.  
   - Se ve reflejado en errores grandes tanto en train como en test.

2. **Overfitting**  
   - Ocurre cuando el modelo es excesivamente complejo y “memoriza” el ruido del conjunto de entrenamiento.  
   - Alta **varianza**: el modelo ajusta muy bien el conjunto de entrenamiento (error bajo), pero falla al generalizar a datos nuevos (error alto en test).  
   - Se ve reflejado en un gap grande entre el error de entrenamiento y el error de validación/prueba.

3. **Curvas de Aprendizaje**  
   - Graficar el error de entrenamiento y validación en función del tamaño de entrenamiento o de la complejidad del modelo (p. ej., profundidad de un árbol).  
   - Permiten diagnosticar si hay overfitting (brecha grande) o underfitting (errores altos en ambos conjuntos).

4. **Técnicas para evitar Overfitting**  
   - **Regularización**: L1 (Lasso), L2 (Ridge) en modelos lineales.  
   - **Pruning** en árboles de decisión.  
   - **Ensamblados**: Random Forest (reduce varianza), Boosting controlado (limitar profundidad de árboles).  
   - **Dropout** en redes neuronales.  
   - **Early stopping**: detener entrenamiento cuando la métrica en validación empeora.

Petal Length (cm) (Longitud del pétalo en cm): Es la longitud de los pétalos de la flor. Los pétalos son la parte colorida de la flor que rodea los órganos reproductivos. Rango típico en el dataset: 1.0 a 6.9 cm.

Petal Width (cm) (Anchura del pétalo en cm): Es la medida del ancho de los pétalos de la flor. Rango típico en el dataset: 0.1 a 2.5 cm.

Target (Objetivo o clase): Esta columna indica a qué especie de flor pertenece el conjunto de medidas de sépalos y pétalos. Los valores son categóricos: 0: Iris Setosa. 1: Iris Versicolor. 2: Iris Virginica.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
````

````python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
````

````python
df['target'] = iris.target
df
````

````python
display(df.head())
print("Proporción de clases:")
print(df['target'].value_counts(normalize=True).rename_axis('clase').reset_index(name='proporción'))
````

````python
sns.pairplot(df, hue='target', diag_kind='kde', palette='Set2')
plt.suptitle("Relaciones entre características según especie", y=1.02)
plt.show()
````

````python
# Calcular promedios por especie
mean_values = df.groupby('target').mean()

# Crear categorías
categories = mean_values.columns
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]

# Plot
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'polar': True})
for species in mean_values.index:
    values = mean_values.loc[species].tolist()
    values += values[:1]  # Cerrar el círculo
    ax.plot(angles, values, label=species)
    ax.fill(angles, values, alpha=0.1)

ax.set_yticks([1, 2, 3, 4, 5, 6])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
plt.title("Comparación promedio de características por especie")
plt.legend(loc='upper right')
plt.show()
````

Dividir entrenamiento y testeo

````python
X = df[iris.feature_names]
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y       # Mantiene proporción de cada clase
)
````

Modelo

````python
clf = DecisionTreeClassifier(
    criterion='gini',     # Métrica de impureza: 'gini' o 'entropy'
    max_depth=3,          # Profundidad máxima para evitar sobreajuste
    min_samples_leaf=5,   # Mín. muestras en cada hoja
    random_state=42
)
clf.fit(X_train, y_train)
````

````python
DecisionTreeClassifier?
````

````python
y_pred = clf.predict(X_test)
````

````python
print(f"Accuracy en test: {accuracy_score(y_test, y_pred):.3f}\n")
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))
````

````python
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm, index=iris.target_names, columns=iris.target_names)
print("Matriz de Confusión:")
display(cm_df)
````

````python
plt.figure(figsize=(12, 8))
plot_tree(
    clf,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True,         # Colorea nodos según la clase dominante
    rounded=True,        # Bordes redondeados
    fontsize=10
)
plt.title("Árbol de Decisión (Iris dataset)")
plt.show()
````

````python
# 1) Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# 2) Cargar dataset Wine
wine_raw = load_wine()
df_wine = pd.DataFrame(wine_raw.data, columns=wine_raw.feature_names)
df_wine['target'] = wine_raw.target

# 3) Mostrar primeros registros
display(df_wine.head())

# 4) Estadísticas descriptivas
display(df_wine.describe().round(2))

# 5) Contar proporción de clases
display(df_wine['target'].value_counts(normalize=True).rename_axis('Clase').reset_index(name='Proporción'))

# 6) Gráfico de pares (pairplot) para ver separabilidad en primeras 3 features
sns.pairplot(
    df_wine,
    vars=['alcohol', 'malic_acid', 'ash'],
    hue='target',
    palette='deep',
    corner=True
)
plt.suptitle('Pairplot de algunas features (Wine)', y=1.02)
plt.show()
````

 Dividir en entrenamiento y prueba, escalado

````python


# 1) Definir X e y
X_wine = df_wine[wine_raw.feature_names]
y_wine = df_wine['target']

# 2) División train/test (70 % – 30 %), estratificando por target
X_train_wine, X_test_wine, y_train_wine, y_test_wine = train_test_split(
    X_wine, y_wine,
    test_size=0.3,
    random_state=42,
    stratify=y_wine
)

# 3) Escalado de features (StandardScaler)
scaler_wine = StandardScaler()
X_train_wine_scaled = scaler_wine.fit_transform(X_train_wine)
X_test_wine_scaled  = scaler_wine.transform(X_test_wine)

# 4) Verificar media y sigma en X_train_scaled
medias = pd.DataFrame(X_train_wine_scaled, columns=wine_raw.feature_names).mean().round(3)
sigmas = pd.DataFrame(X_train_wine_scaled, columns=wine_raw.feature_names).std().round(3)
print("Media (aprox 0):")
display(medias)
print("\nDesviación estándar (aprox 1):")
display(sigmas)
````

- **Comparación de matrices de confusión**  
  - Revisar cuántas muestras de cada clase se clasifican correctamente.  
  - Observar si alguna clase se confunde sistemáticamente con otra.


````python
from sklearn.model_selection import learning_curve
import numpy as np

# Función para graficar curva de aprendizaje
def plot_learning_curve(estimator, title, X, y, cv=5, scoring='accuracy', train_sizes=np.linspace(0.1, 1.0, 5)):
    plt.figure(figsize=(6,4))
    plt.title(title)
    plt.xlabel("Tamaño del conjunto de entrenamiento")
    plt.ylabel("Score")

    train_sizes, train_scores, valid_scores = learning_curve(
        estimator, X, y, cv=cv, scoring=scoring, train_sizes=train_sizes, n_jobs=-1
    )
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std  = np.std(train_scores, axis=1)
    valid_scores_mean = np.mean(valid_scores, axis=1)
    valid_scores_std  = np.std(valid_scores, axis=1)

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1, color="r")
    plt.fill_between(train_sizes, valid_scores_mean - valid_scores_std,
                     valid_scores_mean + valid_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Train score")
    plt.plot(train_sizes, valid_scores_mean, 'o-', color="g", label="CV score")
    plt.legend(loc="best")
    plt.grid(True)
    plt.show()

# 1) Curva de aprendizaje para KNN
plot_learning_curve(
    knn_wine,
    "Curva de Aprendizaje: KNN (Wine)",
    X_train_wine_scaled, y_train_wine,
    cv=5
)

# 2) Curva de aprendizaje para SVM
plot_learning_curve(
    svm_wine,
    "Curva de Aprendizaje: SVM (Wine)",
    X_train_wine_scaled, y_train_wine,
    cv=5
)
````

## 4.4 Ejemplo práctico de Regresión Lineal


Objetivo:
- Usar el dataset “Diabetes” de scikit-learn para predecir la progresión cuantitativa de la enfermedad basada en variables médicas (10 features, todas numéricas).
- Realizar preprocesamiento: dividir en train/test, escalar (opcional para Random Forest pero necesario para Regresión Lineal).
- Entrenar un **LinearRegression** y un **RandomForestRegressor**.
- Evaluar métricas (MSE, R²) y comparar resultados.

````python
from sklearn.datasets import load_diabetes

# 1) Cargar dataset Diabetes
diabetes_raw = load_diabetes()
df_diabetes = pd.DataFrame(diabetes_raw.data, columns=diabetes_raw.feature_names)
df_diabetes['target'] = diabetes_raw.target

# 2) Mostrar primeros registros
display(df_diabetes.head())

# 3) Estadísticas descriptivas
display(df_diabetes.describe().round(2))

# 4) Histograma de la variable target
plt.figure(figsize=(6,4))
sns.histplot(df_diabetes['target'], bins=30, kde=True)
plt.title('Distribución de la variable objetivo (progresión)')
plt.xlabel('Progressión')
plt.ylabel('Conteo')
plt.show()
````

````python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# 1) Definir X e y
X_diab = df_diabetes[diabetes_raw.feature_names]
y_diab = df_diabetes['target']

# 2) División train/test (70 % – 30 %)
X_train_diab, X_test_diab, y_train_diab, y_test_diab = train_test_split(
    X_diab, y_diab, test_size=0.3, random_state=42
)

# 3) Escalado (opcional: las features ya vienen estandarizadas, pero ilustramos el paso)
scaler_diab = StandardScaler()
X_train_diab_scaled = scaler_diab.fit_transform(X_train_diab)
X_test_diab_scaled  = scaler_diab.transform(X_test_diab)

# 4) Instanciar y entrenar Regresión Lineal
lr_diab = LinearRegression()
lr_diab.fit(X_train_diab_scaled, y_train_diab)

# 5) Predecir en test
y_pred_lr = lr_diab.predict(X_test_diab_scaled)

# 6) Evaluar métricas
mse_lr = mean_squared_error(y_test_diab, y_pred_lr)
r2_lr  = r2_score(y_test_diab, y_pred_lr)
print(f"Regresión Lineal – MSE: {mse_lr:.2f}, R²: {r2_lr:.3f}")
````

````python
from sklearn.ensemble import RandomForestRegressor

# 1) Instanciar Random Forest Regressor
rf_diab = RandomForestRegressor(
    n_estimators=100,
    max_depth=None,
    random_state=42
)

# 2) Entrenar
rf_diab.fit(X_train_diab, y_train_diab)  # Random Forest no necesita escalado, puede usar X_train_diab directamente

# 3) Predecir en test
y_pred_rf = rf_diab.predict(X_test_diab)

# 4) Evaluar métricas
mse_rf = mean_squared_error(y_test_diab, y_pred_rf)
r2_rf  = r2_score(y_test_diab, y_pred_rf)
print(f"Random Forest Regressor – MSE: {mse_rf:.2f}, R²: {r2_rf:.3f}")
````

````python
# 1) Graficar valores reales vs predichos para ambos modelos
plt.figure(figsize=(12, 5))

# Subplot 1: Regresión Lineal
plt.subplot(1, 2, 1)
plt.scatter(y_test_diab, y_pred_lr, alpha=0.6)
plt.plot([y_test_diab.min(), y_test_diab.max()], [y_test_diab.min(), y_test_diab.max()], 'r--')
plt.title("Regresión Lineal: Real vs Predicho")
plt.xlabel("Valor real")
plt.ylabel("Valor predicho")

# Subplot 2: Random Forest
plt.subplot(1, 2, 2)
plt.scatter(y_test_diab, y_pred_rf, alpha=0.6, color='green')
plt.plot([y_test_diab.min(), y_test_diab.max()], [y_test_diab.min(), y_test_diab.max()], 'r--')
plt.title("Random Forest: Real vs Predicho")
plt.xlabel("Valor real")
plt.ylabel("Valor predicho")

plt.tight_layout()
plt.show()
````

#5: Algoritmos principales de Clasificación y Regresión

## 5.1. Algoritmos de Clasificación Supervisada

### 5.1.1. Regresión Logística (Logistic Regression)



- **Descripción**:  
  - Modelo lineal para clasificación binaria (y = 0 o 1) que estima la probabilidad de la clase mediante la función sigmoide (𝜎).  
  - Para multiclass, scikit-learn aplica una estrategia “one‐vs‐rest” o “multinomial” con softmax.  

- **Fórmula básica**:  
  $$
    P(y=1 \mid x) = \sigma(w^T x + b) = \frac{1}{1 + e^{-(w^T x + b)}}
  $$
  donde \(w\) son los coeficientes (pesos) y \(b\) el sesgo.

- **Ventajas**:  
  - Fácil de interpretar (coeficientes muestran cuánto impacta cada feature en la probabilidad).  
  - Funciona bien cuando la relación entre X y la probabilidad de y es aproximadamente lineal.  
  - Rápido de entrenar y requiere pocos recursos.

- **Desventajas**:  
  - No captura relaciones no lineales a menos que se agreguen manualmente features polinómicas o transformaciones.  
  - Sensible a multicolinealidad (features altamente correlacionadas).  
  - Para problemas muy grandes o muy desequilibrados puede requerir ajuste cuidadoso de hiperparámetros (`C`, penalización L1/L2).

- **Uso típico**:  
  - Clasificación binaria (spam/no‐spam, fraude/no fraude).  
  - Multiclass con “one‐vs‐rest” si no hay demasiadas clases.  

- **Instanciación en scikit-learn**:
  ```python
  from sklearn.linear_model import LogisticRegression

  model = LogisticRegression(
      penalty='l2',         # 'l1' para Lasso, 'l2' para Ridge
      C=1.0,                # Inverso de fuerza de regularización
      solver='lbfgs',       # Adecuado para multiclass
      multi_class='multinomial',
      max_iter=200,
      random_state=42
  )


### 5.1.2 K-Nearest Neighbors (KNN)


- **Descripción**:  
  - Modelo “perezoso” (lazy learning) que no entrena un modelo explícito.  
  - Para clasificar una nueva muestra, calcula la distancia (por ejemplo, Euclídea) a todas las muestras de entrenamiento y asigna la clase mayoritaria entre sus k vecinos más cercanos.

- **Ventajas**:  
  - Muy intuitivo y sin supuestos sobre la distribución de los datos.  
  - No requiere estimación de parámetros en la fase de entrenamiento (solo almacenar datos).  
  - Interpretable: se puede ver qué vecinos influyen en cada predicción.

- **Desventajas**:  
  - Costoso en predicción (para cada nueva muestra recorre todo el dataset).  
  - Sensible a la escala de las features → requiere escalado previo.  
  - Elección de k crítica:  
    - k muy pequeño → alto riesgo de overfitting (ruido).  
    - k muy grande → puede subajustar (tendencia a la clase mayoritaria).

- **Uso típico**:  
  - Problemas con pocos features y datos moderados.  
  - Clasificación de imágenes pequeñas (cuando cada píxel es una feature) si no hay millones de datos.

- **Instanciación en scikit-learn**:
  ```python
  from sklearn.neighbors import KNeighborsClassifier

  model = KNeighborsClassifier(
      n_neighbors=5,        # k
      weights='uniform',    # 'distance' pondera por inverso de distancia
      metric='minkowski',   # Euclídea (p=2) o Manhattan (p=1)
      p=2
  )


### 5.1.3 Máquinas de Vectores de Soporte (SVM)

- **Descripción**:  
  - Busca el **hiperplano** que separa las clases con el **mayor margen** (distancia mínima entre puntos más cercanos a cada lado).  
  - En caso de no ser linealmente separable, usa un **kernel** (RBF, polinómico, lineal, sigmoidal) para mapear datos a un espacio de mayor dimensión donde puedan separarse linealmente.

- **Ventajas**:  
  - Eficaz en espacios de alta dimensión (features > muestras).  
  - Con kernel RBF, captura relaciones no lineales sin tener que especificar transformaciones manuellement e.g., polinomios de alto grado.  
  - Suele ser robusto y con buen desempeño en datasets medianos.

- **Desventajas**:  
  - Costoso computacionalmente para muy grandes volúmenes de datos.  
  - Selección de kernel, `C` (regularización) y `gamma` (en RBF) es crítica y suele requerir validación cruzada.  
  - No fácilmente interpretable (especialmente con kernel no lineal).

- **Uso típico**:  
  - Problemas con features de alta dimensión (texto, imágenes).  
  - Cuando se presume que la separación entre clases es compleja pero no extremadamente grande en número de muestras.

- **Instanciación en scikit-learn**:
  ```python
  from sklearn.svm import SVC

  model = SVC(
      kernel='rbf',    # 'linear', 'poly', 'sigmoid'
      C=1.0,           # Inverso de fuerza de regularización
      gamma='scale',   # O 'auto', o valor float específico
      probability=False,
      random_state=42
  )


### 5.1.4. Árboles de Decisión (Decision Trees)



- **Descripción**:  
  - Modelo no lineal que particiona recursivamente el espacio de features en regiones “puros” (hojas) basándose en criterios de impureza: Gini, Entropía.  
  - Cada nodo decide un umbral en una feature, dividiendo datos en dos ramas.

- **Ventajas**:  
  - Fácil de interpretar: cada rama es una **regla** “if‐else”.  
  - Maneja variables categóricas y numéricas sin mucho preprocesamiento.  
  - No requiere escalado de features.

- **Desventajas**:  
  - Suelen sobreajustar (overfitting) si no se limitan la profundidad o no se “prunea”.  
  - Pueden ser inestables: un pequeño cambio en los datos puede modificar toda la estructura del árbol.  

- **Uso típico**:  
  - Cuando se requiere interpretabilidad (por ejemplo, en entornos regulatorios).  
  - Como base para ensamblados (Random Forest, Gradient Boosting).

- **Instanciación en scikit-learn**:
  ```python
  from sklearn.tree import DecisionTreeClassifier

  model = DecisionTreeClassifier(
      criterion='gini',      # 'gini' o 'entropy'
      max_depth=None,        # Para limitar sobreajuste, p. ej. max_depth=5
      min_samples_split=2,   # Mínimo número de muestras para dividir un nodo
      min_samples_leaf=1,    # Mínimo de muestras para un nodo hoja
      random_state=42
  )



### 5.1.5. Random Forest





- **Descripción**:  
  - Ensamble de varios Árboles de Decisión entrenados con *bagging* (bootstrap aggregation).  
  - Cada árbol se entrena con un **subconjunto aleatorio de muestras** (con reemplazo) y, en cada división, se consideran solo un **subconjunto aleatorio de features**.  
  - Durante la predicción, la clase final se elige por **votación mayoritaria** (clasificación) o por **promedio** de predicciones (regresión).

- **Ventajas**:  
  - Reduce el sobreajuste de árboles individuales (mayor estabilidad y menor varianza).  
  - Maneja bien datos con alto número de features y muestra variabilidad de importancia de variables.  
  - Requiere menos tuning que un solo árbol para obtener buen desempeño.

- **Desventajas**:  
  - Menos interpretables que un árbol único (aunque existe forma de obtener “feature importance”).  
  - Puede volverse pesado en tiempo de inferencia si se usan cientos de árboles.  
  - No modela relaciones extremadamente complejas que Gradient Boosting puede capturar mejor.

- **Uso típico**:  
  - Problemas con datos tabulares medianos a grandes donde la interpretabilidad no es crítica.  
  - Cuando se busca un modelo sólido con tuning mínimo de hiperparámetros (`n_estimators`, `max_depth`).

- **Instanciación en scikit-learn**:
  ```python
  from sklearn.ensemble import RandomForestClassifier

  model = RandomForestClassifier(
      n_estimators=100,        # Número de árboles en el ensamble
      criterion='gini',
      max_depth=None,          # Limitar profundidad para evitar overfitting
      max_features='sqrt',     # Número de features consideradas en cada división
      random_state=42,
      n_jobs=-1                # Usa todos los cores disponibles
  )


### 5.1.6. Gradient Boosting (XGBoost / LightGBM / CatBoost)




- **Descripción**:  
  - Ensamble secuencial de árboles “débiles” (shallow trees) donde cada nuevo árbol corrige los errores (residuos) de la suma de predicciones de los árboles anteriores.  
  - Minimiza una función de pérdida (por ejemplo, log‐loss para clasificación, MSE para regresión) mediante **descenso de gradiente** en espacio de modelos.

- **Implementaciones populares**:  
  - **XGBoost**: eficiente, muy popular en competiciones de ML por su velocidad y manejos de regularización.  
  - **LightGBM**: optimizado para datasets muy grandes; utiliza histogram-based splitting para velocidad.  
  - **CatBoost**: maneja variables categóricas de forma nativa sin necesidad de “one-hot encoding”.

- **Ventajas**:  
  - Data-modeling muy poderoso: captura relaciones no lineales y complejas.  
  - Suelen ofrecer mejor performance en competiciones y problemas reales.  
  - Manejan bien características categóricas (especialmente CatBoost) y datos con missing values (XGBoost y LightGBM pueden manejarlos).

- **Desventajas**:  
  - Mayor complejidad y tiempo de entrenamiento (comparado con Random Forest).  
  - Requiere ajuste cuidadoso de hiperparámetros (`learning_rate`, `n_estimators`, `max_depth`, `subsample`).  
  - Menos interpretables; aunque existen librerías que generan “SHAP values” para explicabilidad.

- **Uso típico**:  
  - Competencias de Kaggle / problemas reales donde se busca máxima precisión.  
  - Datasets tabulares con variables heterogéneas (numéricas y categóricas).

- **Instanciación ejemplo con XGBoost**:
  ```python
  from xgboost import XGBClassifier

  model = XGBClassifier(
      n_estimators=100,
      learning_rate=0.1,
      max_depth=6,
      subsample=0.8,
      colsample_bytree=0.8,
      random_state=42,
      use_label_encoder=False,
      eval_metric='mlogloss'
  )


### 5.1.7. Redes Neuronales Básicas (MLPClassifier)



- **Descripción**:  
  - Perceptrón multicapa (Multi‐Layer Perceptron) con al menos una capa oculta.  
  - Cada neurona se conecta a todas las neuronas de la capa siguiente, usando una función de activación (ReLU, tanh, sigmoide).  
  - Se entrena con **backpropagation** para ajustar pesos, minimizando una función de pérdida (cross‐entropy para clasificación).

- **Ventajas**:  
  - Capacidad de modelar relaciones muy complejas no lineales.  
  - Con suficientes datos y capas, puede aproximar casi cualquier función (teorema de aproximación universal).  
  - Ofrece flexibilidad en arquitectura: número de capas, neuronas por capa, funciones de activación.

- **Desventajas**:  
  - Requiere más datos y tuning de hiperparámetros (número de capas, neuronas, tasa de aprendizaje, regularización).  
  - Entrenamiento más lento que modelos clásicos, especialmente con grandes redes.  
  - Menor interpretabilidad: es difícil explicar por qué toma cierta decisión.

- **Uso típico**:  
  - Problemas complejos de clasificación con muchos datos (por ejemplo, imágenes, texto bag‐of‐words).  
  - Cuando otros algoritmos tradicionales no logran buena precisión.

- **Instanciación en scikit-learn**:
  ```python
  from sklearn.neural_network import MLPClassifier

  model = MLPClassifier(
      hidden_layer_sizes=(100,),  # Una capa oculta con 100 neuronas
      activation='relu',          # Función de activación: 'relu', 'tanh', 'logistic'
      solver='adam',              # Optimizador: 'adam', 'sgd'
      alpha=0.0001,               # Tasa de regularización L2
      learning_rate='constant',   # O 'adaptive'
      max_iter=200,
      random_state=42
  )


### 5.1.8. Métricas de Evaluación en Clasificación



- **Accuracy**:  
  $$
    \text{Accuracy} = \frac{\text{TP} + \text{TN}}{\text{Total de muestras}}
  $$  
  Es la proporción de predicciones correctas.  
  *No recomendable cuando hay clases desbalanceadas.*

- **Precision** (por clase):  
  $$
    \text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}}
  $$  
  Proporción de verdaderos positivos sobre todas las predicciones positivas.  
  *Indica qué tan “confiables” son las predicciones positivas.*

- **Recall** (por clase) o Sensibilidad:  
  $$
    \text{Recall} = \frac{\text{TP}}{\text{TP} + \text{FN}}
 $$  
  Proporción de verdaderos positivos sobre todos los positivos reales.  
  *Mide la capacidad de detectar todos los positivos reales.*

- **F1-score**:  
  $$
    F1 = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
  $$  
  Promedio armónico entre precision y recall.  
  *Equilibra ambos, útil cuando hay desbalance de clases.*

- **Matriz de Confusión**:  
  - Tabla que muestra conteos de TP, FP, FN y TN por clase (en multiclass es matriz con filas = etiquetas reales, columnas = predichas).  
  - Permite ver qué clases se confunden entre sí.

- **ROC‐AUC** (solo binaria o “ovr” en multiclass):  
  - **ROC curve**: gráfica de TPR (recall) vs FPR (falsos positivos) para distintos umbrales.  
  - **AUC**: área bajo la curva ROC.  
  - Valores cercanos a 1 indican buen separador; 0.5 indica azar.


## 5.2. Algoritmos de Regresión Supervisada

### 5.2.1. Regresión Lineal (Linear Regression)



- **Descripción**:  
  - Modelo lineal que ajusta un hiperplano \(\displaystyle \hat{y} = w^T x + b\) minimizando la suma de cuadrados de los residuos (Ordinary Least Squares).  
  - Fácil de interpretar: cada coeficiente \(w_i\) indica cambio esperado en \(\hat{y}\) cuando la feature \(x_i\) aumenta en 1 unidad, manteniendo otras constantes.

- **Ventajas**:  
  - Simple y rápido de entrenar.  
  - Interpretabilidad directa de coeficientes.  
  - Base para entender conceptos de optimización y métricas (MSE, R²).

- **Desventajas**:  
  - Asume **linearidad** entre features y target.  
  - Sensible a outliers (pueden sesgar los coeficientes).  
  - No maneja multicolinealidad sin regularización.

- **Uso típico**:  
  - Problemas donde la relación aproximada entre X e y es lineal o se puede linearizar.  
  - Como baseline antes de intentar modelos más complejos.

- **Instanciación en scikit-learn**:
  ```python
  from sklearn.linear_model import LinearRegression

  model = LinearRegression(
      fit_intercept=True,
      normalize=False  # En versiones recientes, se depreca. Mejor usar StandardScaler si se quiere normalizar.
  )


### 5.2.2. Regresión Polinómica (Polynomial Regression)



- **Descripción**:  
  - Extiende la regresión lineal añadiendo **features polinómicas** (p. ej., \(x^2, x^3\), interacciones).  
  - Permite modelar relaciones no lineales en forma de polinomio.  
  - Se construye un pipeline:  
    1. `PolynomialFeatures(degree=d)`: genera columnas \([x, x^2, x^3, \ldots]\).  
    2. `LinearRegression()`: ajusta coeficientes a esas nuevas features.

- **Ventajas**:  
  - Mantiene interpretabilidad (todavía es modelo lineal en las features transformadas).  
  - Captura cierta no linealidad sin recurrir a modelos muy complejos.

- **Desventajas**:  
  - Riesgo de sobreajuste si el grado del polinomio es muy alto.  
  - Número de features crece exponencialmente con el grado y el número de variables originales.  

- **Uso típico**:  
  - Cuando se observa curvatura en la relación X vs y.  
  - En problemas con pocas variables, se puede probar con grado 2 o 3 antes de pasar a modelos más complejos.

- **Instanciación / Ejemplo en scikit-learn**:
  ```python
  from sklearn.preprocessing import PolynomialFeatures
  from sklearn.linear_model import LinearRegression
  from sklearn.pipeline import Pipeline

  model = Pipeline([
      ('poly', PolynomialFeatures(degree=2, include_bias=False)),
      ('linreg', LinearRegression())
  ])


### 5.2.3. K-Nearest Neighbors Regressor



- **Descripción**:  
  - Similar a KNN de clasificación, pero la predicción es el **promedio** (o mediana) de los valores de las k muestras más cercanas en el espacio de features.  
  - No construye un modelo explícito; es un método “perezoso”.

- **Ventajas**:  
  - Intuitivo: no asume forma paramétrica para la relación.  
  - Fácil de implementar y entender.

- **Desventajas**:  
  - Sensible a la escala → requiere escalado.  
  - La calidad de la predicción depende de la densidad de muestras cercanas: si hay pocas muestras similares, el promedio puede no ser representativo.  
  - Costoso en predicción: para cada punto nuevo, calcula distancia a todos los datos de entrenamiento.

- **Uso típico**:  
  - Problemas con datasets medianos y pocas dimensiones.  
  - Situaciones donde se prefiere un modelo no paramétrico sin suposiciones de forma.

- **Instanciación en scikit-learn**:
  ```python
  from sklearn.neighbors import KNeighborsRegressor

  model = KNeighborsRegressor(
      n_neighbors=5,
      weights='uniform',      # O 'distance'
      metric='minkowski',     # Euclidean (p=2) o Manhattan (p=1)
      p=2
  )


### 5.2.4. Support Vector Regressor (SVR)



- **Descripción**:  
  - Extiende el concepto de SVM a regresión.  
  - En lugar de maximizar el margen entre clases, SVR define un “epsilon‐tube” donde los errores no se penalizan si quedan dentro de \(\varepsilon\).  
  - Minimiza tanto la complejidad del modelo (norma de pesos) como la suma de errores fuera del epsilon­-tube.

- **Ventajas**:  
  - Captura relaciones lineales y no lineales (usando kernels: RBF, polinómico).  
  - Robusto a outliers hasta cierto punto (por el epsilon).

- **Desventajas**:  
  - Requiere ajuste de hiperparámetros: `C` (regularización), `epsilon`, `kernel`, `gamma`.  
  - Costoso en tiempo si el dataset es grande.

- **Uso típico**:  
  - Problemas de regresión con dimensión moderada y relaciones no lineales.  
  - Situaciones donde se necesita un modelo robusto con buen control de errores.

- **Instanciación en scikit-learn**:
  ```python
  from sklearn.svm import SVR

  model = SVR(
      kernel='rbf',     # 'linear', 'poly'
      C=1.0,
      epsilon=0.1,      # Tamaño del epsilon‐tube
      gamma='scale'
  )


### 5.2.5. Árboles de Regresión (DecisionTreeRegressor)



- **Descripción**:  
  - Similar a los árboles de decisión para clasificación, pero en las hojas se predice el **promedio** (o mediana) de los valores de la variable target dentro de la región.  
  - Divide recursivamente el espacio de features minimizando la MSE (o MAE) en cada partición.

- **Ventajas**:  
  - No lineal: captura interacciones y relaciones complejas sin transformaciones previas.  
  - No requiere escalado de features.  
  - Fácil de visualizar e interpretar (diagramas de árbol).

- **Desventajas**:  
  - Muy propenso a sobreajustar si no se limita la profundidad (`max_depth`) o el número mínimo de muestras por hoja (`min_samples_leaf`).  
  - Inestable: pequeño cambio en datos puede cambiar la forma del árbol.

- **Uso típico**:  
  - Cuando se busca un modelo interpretable para regresión.  
  - Como base para ensambles (Random Forest Regressor, Gradient Boosting Regressor).

- **Instanciación en scikit-learn**:
  ```python
  from sklearn.tree import DecisionTreeRegressor

  model = DecisionTreeRegressor(
      criterion='mse',        # o 'friedman_mse', 'mae'
      max_depth=None,         # Limitar para evitar overfitting
      min_samples_split=2,
      min_samples_leaf=1,
      random_state=42
  )


### 5.2.6. Random Forest Regressor



- **Descripción**:  
  - Ensamble de árboles de regresión entrenados con *bagging* y selección aleatoria de features por división.  
  - La predicción final es el **promedio** de las predicciones de cada árbol.

- **Ventajas**:  
  - Reduce varianza comparado con un solo árbol → menor sobreajuste.  
  - Maneja relaciones no lineales y altas dimensiones.  
  - Importancia de features fácilmente extraíble (`feature_importances_`).

- **Desventajas**:  
  - Menos interpretable que un solo árbol.  
  - Requiere más memoria y tiempo en predicción si se usan muchos árboles.

- **Uso típico**:  
  - Problemas de regresión tabular donde se busca un modelo robusto sin mucho tuning.  
  - Cuando hay muchas features irrelevantes, RF tiende a ignorarlas por “bagging” y selección aleatoria.

- **Instanciación en scikit-learn**:
  ```python
  from sklearn.ensemble import RandomForestRegressor

  model = RandomForestRegressor(
      n_estimators=100,
      criterion='mse',
      max_depth=None,
      max_features='auto',    # O 'sqrt', 'log2'
      random_state=42,
      n_jobs=-1
  )


### 5.2.7. Gradient Boosting Regressor



- **Descripción**:  
  - Ensamble secuencial de árboles de regresión que corrigen errores de iteraciones previas.  
  - Minimiza la función de pérdida (MSE, MAE) mediante descenso de gradiente en el espacio de modelos.

- **Implementaciones populares**:  
  - **XGBoostRegressor** (XGBoost para regresión).  
  - **LGBMRegressor** (LightGBM para regresión).  
  - **CatBoostRegressor** (CatBoost para regresión).

- **Ventajas**:  
  - Alto poder predictivo en relaciones no lineales complejas.  
  - Control fino de sobreajuste a través de hiperparámetros (`learning_rate`, `n_estimators`, `max_depth`, `subsample`).  
  - Manejo nativo de valores faltantes (XGBoost, LightGBM) y variables categóricas (CatBoost).

- **Desventajas**:  
  - Más lento de entrenar que Random Forest si el número de árboles y su complejidad es alto.  
  - Requiere tuning exhaustivo de hiperparámetros para obtener el mejor rendimiento.

- **Instanciación ejemplo con XGBoost**:
  ```python
  from xgboost import XGBRegressor

  model = XGBRegressor(
      n_estimators=100,
      learning_rate=0.1,
      max_depth=6,
      subsample=0.8,
      colsample_bytree=0.8,
      random_state=42,
      objective='reg:squarederror'
  )


### 5.2.8. Regresión Regularizada (Lasso, Ridge, ElasticNet)



- **Lasso (L1)**:  
  - Penaliza la suma de valores absolutos de coeficientes: $\alpha \sum |w_i|$.  
  - Tiende a producir coeficientes exactamente 0 → selección automática de features.

- **Ridge (L2)**:  
  - Penaliza la suma de cuadrados de coeficientes: $\alpha \sum w_i^2$.  
  - No elimina coeficientes pero los “encoge” hacia 0 para reducir varianza.

- **ElasticNet**:  
  - Combinación de L1 y L2: $\alpha (\rho \sum |w_i| + (1-\rho)\sum w_i^2)$.  
  - Equilibrio entre selección de features (Lasso) y estabilidad (Ridge).

- **Ventajas**:  
  - Reducen overfitting en regresión lineal cuando hay muchas features o multicolinealidad.  
  - Lasso realiza selección de variables implícita.  

- **Desventajas**:  
  - Requieren ajuste de parámetro $\alpha$ (grado de penalización) y, en ElasticNet, también $\rho$.  
  - Si $\alpha$ es muy alto, el modelo puede subajustar.

- **Instanciación en scikit-learn**:
  - **Lasso**:
    ```python
    from sklearn.linear_model import Lasso

    model = Lasso(
        alpha=1.0,        # Grado de penalización L1
        max_iter=1000,
        random_state=42
    )
    ```
  - **Ridge**:
    ```python
    from sklearn.linear_model import Ridge

    model = Ridge(
        alpha=1.0,        # Grado de penalización L2
        max_iter=1000,
        random_state=42
    )
    ```
  - **ElasticNet**:
    ```python
    from sklearn.linear_model import ElasticNet

    model = ElasticNet(
        alpha=1.0,        # Grado de penalización total
        l1_ratio=0.5,     # ρ: proporción de L1 vs L2
        max_iter=1000,
        random_state=42
    )
    ```


#### 5.2.9. Métricas de Evaluación en Regresión



- **Mean Absolute Error (MAE)**:  
$$
    \text{MAE} = \frac{1}{n}\sum_{i=1}^n |y_i - \hat{y}_i|
$$  
  Promedio de errores absolutos.  
  *Fácil de interpretar (misma unidad que y).*

- **Mean Squared Error (MSE)**:  
  $$
    \text{MSE} = \frac{1}{n}\sum_{i=1}^n (y_i - \hat{y}_i)^2
  $$  
  Penaliza más los errores grandes (al elevar al cuadrado).

- **Root Mean Squared Error (RMSE)**:  
  $$
    \text{RMSE} = \sqrt{\text{MSE}}
  $  
  Misma unidad que y, facilita interpretación.

- **R² (Coeficiente de determinación)**:  
  $$
    R^2 = 1 - \frac{\sum_{i=1}^n (y_i - \hat{y}_i)^2}{\sum_{i=1}^n (y_i - \bar{y})^2}
  $$  
  Mide la proporción de varianza de y explicada por el modelo.  
  - $R^2 = 1$: ajuste perfecto.  
  - $R^2 = 0$: el modelo no explica nada más que la media.  
  - Puede ser negativo si el modelo ajusta peor que la media.

- **Adjusted R²**:  
  $$
    \bar{R}^2 = 1 - \frac{(1 - R^2)(n - 1)}{n - p - 1}
  $$  
  Ajusta R² penalizando el número de features $(p)$ para evitar sobreestimación cuando hay muchas variables.

- **Explained Variance Score**:  
  $$
    \text{EV} = 1 - \frac{\text{Var}(y - \hat{y})}{\text{Var}(y)}
  $$  
  Similar a R², mide qué fracción de la varianza total se explica.

---

**Nota**:  
- En problemas de regresión, conviene reportar al menos **MSE (o RMSE)** y **R²**.  
- A veces se incluye **MAE** si es crucial interpretar el error en la misma unidad que la variable.  
- Para comparar modelos con diferentes variables objetivo, usar métricas **relativas** (por ejemplo, MAPE: Mean Absolute Percentage Error).

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 6) Instanciar Random Forest Classifier
rf_clf = RandomForestClassifier(
    n_estimators=100,
    criterion='gini',
    max_depth=None,
    max_features='sqrt',
    random_state=42,
    n_jobs=-1
)

# 7) Entrenar sobre datos escalados
rf_clf.fit(X_train_wine_scaled, y_train_wine)

# 8) Predecir etiquetas en el conjunto de prueba
y_pred_wine = rf_clf.predict(X_test_wine_scaled)

# 9) Evaluar métricas
acc_wine = accuracy_score(y_test_wine, y_pred_wine)
print(f"Accuracy (Wine – Random Forest): {acc_wine:.3f}\n")

print("Classification Report (Wine – Random Forest):")
print(classification_report(y_test_wine, y_pred_wine, target_names=wine_raw.target_names))

# 10) Matriz de confusión
cm_wine = confusion_matrix(y_test_wine, y_pred_wine)
cm_wine_df = pd.DataFrame(cm_wine, index=wine_raw.target_names, columns=wine_raw.target_names)
print("Matriz de Confusión (Wine – Random Forest):")
display(cm_wine_df)
````

````python
# 11) Extraer importancia de cada feature
importancias = rf_clf.feature_importances_
feat_importance_df = pd.DataFrame({
    'feature': wine_raw.feature_names,
    'importance': importancias
}).sort_values(by='importance', ascending=False)

# 12) Mostrar tabla ordenada
print("Importancia de features (Wine):")
display(feat_importance_df.reset_index(drop=True))

# 13) Graficar las 8 features más importantes
plt.figure(figsize=(8,5))
sns.barplot(
    x='importance',
    y='feature',
    data=feat_importance_df.head(8),
    palette='viridis'
)
plt.title('Top 8 Features más importantes (Random Forest – Wine)')
plt.xlabel('Importancia')
plt.ylabel('Feature')
plt.show()
````

# Ejemplo A: Regresión Lineal (Diabetes)

## 1. ¿Qué es la Regresión Lineal?
La regresión lineal es una técnica estadística y de Machine Learning que permite **predecir un valor numérico** (por ejemplo, cuánto empeora una enfermedad) a partir de una o más variables conocidas (por ejemplo, el índice de masa corporal, BMI).  
En su forma más simple (regresión lineal **univariable**), queremos ajustar una **línea** que relaciona una sola variable \(x\) con el valor que queremos predecir \(y\). Matemáticamente, esa línea se escribe así:

$$
\hat{y} \;=\; w \cdot x \;+\; b
$$

- $x$ es la característica (por ejemplo, “BMI”).  
- $w$ es la **pendiente** de la línea (cuánto cambia $\hat{y}$ cuando $x$ aumenta en 1).  
- \(b\) es el **intercepto** (el valor de $\hat{y}$ cuando $x=0$).  
- \(\hat{y}\) (se lee “y sombrero”) es la **predicción** que hace el modelo para cada \(x\).

### 1.1 ¿Por qué una línea?
Supongamos que tenemos datos de varios pacientes, cada uno con su valor de BMI ($x_i$) y un valor de “progresión de la enfermedad” ($y_i$). Queremos encontrar una línea que pase lo más cerca posible de todos esos puntos $\bigl(x_i,\,y_i\bigr)$. De esa forma, si recibimos un nuevo paciente con cierto BMI, podemos calcular  
$$
\hat{y} = w \cdot x_{\text{nuevo}} + b
$$
y dar una estimación de su progresión de la enfermedad.

---

## 2. ¿Cómo encontramos \(w\) y \(b\)?
Para “encajar” la mejor línea, medimos qué tan lejos están las predicciones \(\hat{y}_i\) de los valores reales \(y_i\). La forma estándar de medir este error es la **Media de los Errores Cuadrados** (MSE, Mean Squared Error):

\[
\mathrm{MSE}(w,b)
\;=\; \frac{1}{n} \sum_{i=1}^n \bigl(y_i \;-\; \hat{y}_i\bigr)^{2}
\;=\; \frac{1}{n} \sum_{i=1}^n \bigl(y_i \;-\; (\,w\,x_i + b\,)\bigr)^{2},
\]

donde:
- \(n\) es el número total de muestras (pacientes) en el conjunto de entrenamiento.  
- \(y_i\) es el valor real de progresión para la muestra \(i\).  
- \(\hat{y}_i = w\,x_i + b\) es la predicción para la muestra \(i\).

### 2.1. Minimizar el MSE
Queremos encontrar los parámetros \(w\) y \(b\) que **minimicen** \(\mathrm{MSE}(w,b)\). Existen dos enfoques principales:

1. **Solución analítica (Normal Equation)**  
   Para regresión lineal univariable (una sola característica), podemos derivar fórmulas cerradas:
   \[
   w = \frac{\displaystyle \sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{\displaystyle \sum_{i=1}^n (x_i - \bar{x})^2},
   \quad
   b = \bar{y} - w\,\bar{x},
   \]
   donde \(\bar{x}\) y \(\bar{y}\) son las medias de \(x_i\) y \(y_i\).

2. **Descenso por Gradiente (Gradient Descent)**  
   Para muchos datos o varias características, iteramos ajustando \(w\) y \(b\) en pequeños pasos que disminuyen el MSE.  
   En cada iteración:
   \[
   w \;\leftarrow\; w \;-\; \alpha \,\frac{\partial \mathrm{MSE}}{\partial w},
   \quad
   b \;\leftarrow\; b \;-\; \alpha \,\frac{\partial \mathrm{MSE}}{\partial b},
   \]
   donde \(\alpha\) es la tasa de aprendizaje (learning rate), y los gradientes son:
   \[
   \frac{\partial \mathrm{MSE}}{\partial w}
   = -\frac{2}{n} \sum_{i=1}^n x_i \bigl(y_i - (w\,x_i + b)\bigr),
   \quad
   \frac{\partial \mathrm{MSE}}{\partial b}
   = -\frac{2}{n} \sum_{i=1}^n \bigl(y_i - (w\,x_i + b)\bigr).
   \]
   Con suficientes iteraciones, converge a valores de \(w\) y \(b\) cercanos a los óptimos.

En bibliotecas como **scikit-learn**, usamos internamente optimizadores eficientes (o resolución directa) para calcular \(w\) y \(b\). Nosotros solo llamamos a `LinearRegression().fit(X, y)`.

---

## 3. Métrica de Rendimiento: R²
Además del MSE, se usa el coeficiente de determinación \(R^2\), que mide qué proporción de la varianza de \(y\) explica el modelo. Se define como:

$$R^2
\;=\; 1 \;-\;
\frac{\displaystyle \sum_{i=1}^n \bigl(y_i - \hat{y}_i\bigr)^2}
     {\displaystyle \sum_{i=1}^n \bigl(y_i - \bar{y}\bigr)^2}.
$$

- Si $R^2 = 1$, el modelo predice perfectamente todos los puntos.  
- Si $R^2 = 0$, el modelo no explica nada mejor que predecir siempre la media $\bar{y}$.  
- Si $R^2$ es negativo, el modelo rinde peor que predecir la media.

---

## 4. Interpretación Gráfica: Real vs Predicho
Para visualizar cómo de bien ajusta la línea, graficamos en un diagrama de dispersión:

- Eje X: valores reales \(y_i\).  
- Eje Y: valores predichos \(\hat{y}_i\).

Y dibujamos la **línea de referencia** $y = \hat{y}$ (línea punteada).  
- Si la mayoría de los puntos se agrupa muy cerca de esa línea, el modelo ajusta muy bien.  
- Si hay muchos puntos alejados de la línea, el ajuste es más pobre.

---

## 5. Visión Intuitiva para Estudiantes
- Imagina que tienes varios puntos \((x_i, y_i)\) en un plano.  
- Queremos “dibujar” una línea recta que pase lo más cerca posible de todos esos puntos.  
- Al hacerlo, si nos dan un nuevo \(x\), usamos esa línea para calcular \(\hat{y}\).  
- Cuanto más “pegada” esté la línea a los puntos de entrenamiento, mejor será la predicción.

## 7. Ejemplo Interactivo: Predicción a partir de un valor de BMI

Para que los estudiantes vean en tiempo real cómo cambia la predicción al ingresar un nuevo valor de BMI, podemos usar un widget interactivo. En Google Colab (o Jupyter) podemos emplear **ipywidgets** para esto. A continuación:

1. Mostramos de nuevo el entrenamiento del modelo (ya aprendido).  
2. Creamos un **slider** o cuadro de texto para que el estudiante introduzca un valor de BMI (entre el mínimo y el máximo de nuestro dataset).  
3. Cada vez que cambie ese valor, el modelo calculará la predicción y la imprimirá al instante.

De este modo, al mover el slider o escribir un número, el estudiante verá cómo la regresión lineal traduce ese BMI en un valor estimado de progresión de la enfermedad.


````python
# 7.1) Importar librerías y volver a entrenar (si no está en memoria)
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import ipywidgets as widgets
from IPython.display import display

# -- Cargar datos y armar DataFrame --
data = load_diabetes()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# -- Seleccionar solo la característica 'bmi' --
X = df[['bmi']]
y = df['target']

# -- Dividir en entrenamiento y prueba --
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# -- Entrenar regresión lineal con la feature 'bmi' --
model = LinearRegression()
model.fit(X_train, y_train)

# 7.2) Crear widget para que el estudiante ingrese el valor de BMI
bmi_slider = widgets.FloatSlider(
    min=float(df['bmi'].min()),
    max=float(df['bmi'].max()),
    step=0.01,
    description='BMI:',
    continuous_update=False
)

# 7.3) Área de salida donde mostraremos la predicción
output = widgets.Output()

# 7.4) Función que se ejecuta cada vez que cambia el slider
def predict_progression(bmi_value):
    with output:
        output.clear_output()               # Limpiar salida anterior
        pred = model.predict([[bmi_value]])[0]
        print(f"Valor de BMI seleccionado: {bmi_value:.2f}")
        print(f"Predicción de progresión de la enfermedad: {pred:.2f}")

# 7.5) Vincular la función al slider
widgets.interact(predict_progression, bmi_value=bmi_slider)

# 7.6) Mostrar el área de salida
display(output)
````

````python
import pandas as pd

# Para cargar el dataset
from sklearn.datasets import load_iris

from sklearn.model_selection import train_test_split

# Para escalar (StandardScaler) y entrenar Regresión Logística
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Métricas de evaluación
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Para graficar
import matplotlib.pyplot as plt
import seaborn as sns

# (Opcional) fijar estilo de Seaborn
sns.set(style="whitegrid", palette="muted")

````

````python
iris_raw = load_iris()
iris_raw
````

````python
# iris_raw contiene:
# - iris_raw.data: matriz (n_samples x n_features)
# - iris_raw.target: vector de etiquetas codificadas 0,1,2
# - iris_raw.feature_names: nombres de las columnas (cuatro mediciones)
# - iris_raw.target_names: nombres de cada clase: ['setosa', 'versicolor', 'virginica']

# Convertir iris_raw.data en un DataFrame
df_iris = pd.DataFrame(iris_raw.data, columns=iris_raw.feature_names)
df_iris

````

Variable objetivo

````python
# 3Agregar la columna target numérica
df_iris['target_num'] = iris_raw.target
df_iris
````

````python

# Crear columna con etiqueta en texto (más legible)
df_iris['species'] = df_iris['target_num'].map(
    {i: name for i, name in enumerate(iris_raw.target_names)}
)
df_iris
````

````python
# Mostrar las primeras 5 filas para verificar
df_iris.head()
````

EDA

````python
# 1) Descripción estadística básica
df_iris.describe().round(2)
````

````python
# 2) Gráfico de dispersión: petal length vs petal width, coloreado por especie
plt.figure(figsize=(6, 4))
sns.scatterplot(
    data=df_iris,
    x='petal length (cm)',
    y='petal width (cm)',
    hue='species',
    palette='deep'
)
plt.title('Petal Length vs Petal Width (Iris)')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Especie')
plt.show()
````

````python
# 3) Histograma de “sepal length” por especie
plt.figure(figsize=(6, 4))
sns.histplot(
    data=df_iris,
    x='sepal length (cm)',
    hue='species',
    multiple='stack',
    bins=15
)
plt.title('Distribución de Sepal Length por Especie')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Conteo')
plt.show()
````

````python
# 1) Definir X e y
X = df_iris[iris_raw.feature_names]    # solo las 4 columnas numéricas
y = df_iris['target_num']              # etiqueta en números (0,1,2)

# 2) División en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

# 3) Mostrar tamaños de cada subconjunto
print(f"Total de muestras: {X.shape[0]}")
print(f"  Entrenamiento: {X_train.shape[0]} muestras")
print(f"  Prueba:       {X_test.shape[0]} muestras")
````

 Escalar características (StandardScaler)

Aunque Regresión Logística no exige escala para funcionar, es buena práctica estandarizar para que los coeficientes sean comparables y acelerar convergencia.

En problemas con KNN, SVM, etc., es imprescindible que X tenga media 0 y varianza 1.

````python
# 1) Instanciar el escalador
scaler = StandardScaler()

# 2) Ajustar el escalador solo con datos de entrenamiento
scaler.fit(X_train)

# 3) Transformar (escalar) tanto X_train como X_test
X_train_scaled = scaler.transform(X_train)
X_test_scaled  = scaler.transform(X_test)

# 4) Verificar media y varianza del X_train_scaled
print("Media de cada feature en X_train_scaled (debe ser ~0):")
print(pd.DataFrame(X_train_scaled, columns=iris_raw.feature_names).mean().round(3))

print("\nDesviación estándar en X_train_scaled (debe ser ~1):")
print(pd.DataFrame(X_train_scaled, columns=iris_raw.feature_names).std().round(3))
````

Elegir e instanciar un modelo (Regresión Logística)

````python
# 1) Instanciar el modelo de regresión logística
model = LogisticRegression(
    multi_class='multinomial',
    solver='lbfgs',
    max_iter=200,
    random_state=42
)

# 2) Revisar los hiperparámetros por defecto
model
````

Entrenar el modelo (fit) con los datos de entrenamiento
Descripción:

El método fit(X_train_scaled, y_train) ajusta los coeficientes internos mediante descenso de gradiente (o LBFGS), minimizando la función de pérdida log-loss.

Una vez entrenado, model.coef_ y model.intercept_ representan los parámetros internos.

````python
# 1) Entrenamiento
model.fit(X_train_scaled, y_train)

# 2) Mostrar coeficientes del modelo (breve inspección)
print("Coeficientes (por clase):")
print(pd.DataFrame(
    model.coef_,
    index=iris_raw.target_names,
    columns=iris_raw.feature_names
).round(3))

print("\nIntercepts (bias de cada clase):")
print(pd.Series(model.intercept_, index=iris_raw.target_names).round(3))
````

model.coef_ es un array de forma (n_classes, n_features), donde cada fila indica los pesos aprendidos para esa clase.

model.intercept_ es un vector de sesgos (bias) para cada clase.

Podemos ver que, para setosa, el peso asociado a “petal length” es muy negativo (indica que valores grandes de petal length alejan la probabilidad de ser setosa).

Estas interpretaciones son aproximadas, pero ayudan a entender la contribución de cada feature en la probabilidad de pertenecer a cierta clase.

Paso 8: Predecir sobre datos de prueba
Descripción:

Con predict(X_test_scaled) obtenemos una etiqueta predicha (0,1 o 2) para cada muestra de prueba.

Con predict_proba(X_test_scaled) podemos ver la probabilidad estimada de cada clase (opcional).

````python
# 1) Predecir etiquetas
y_pred = model.predict(X_test_scaled)

# 2) Opcional: Predecir probabilidades de cada clase
y_probas = model.predict_proba(X_test_scaled)

# 3) Mostrar las primeras 5 filas de probabilidades
pd.DataFrame(
    y_probas,
    columns=iris_raw.target_names
).head()
````

Cada fila muestra la probabilidad de que esa muestra pertenezca a cada una de las tres clases.

El método predict() asigna la etiqueta con probabilidad más alta.

Paso 9: Evaluar desempeño del modelo
Descripción:

Calculamos la accuracy global: proporción de aciertos sobre el total de muestras de prueba.

Con classification_report obtenemos métricas más completas: precision, recall y F1‐score por clase.

También generamos la matriz de confusión para ver cuántas muestras de cada clase se confundieron.

````python
# 1) Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy (conjunto de prueba): {accuracy:.3f}")

# 2) Classification report (precision, recall, f1-score)
print("\nClassification Report:")
print(classification_report(
    y_test,
    y_pred,
    target_names=iris_raw.target_names
))

# 3) Matriz de confusión
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(
    cm,
    index=iris_raw.target_names,
    columns=iris_raw.target_names
)

print("\nMatriz de Confusión:")
print(cm_df)
````

Accuracy = 0.978 indica que, de 45 muestras de prueba, solo se equivocó en 1 (probablemente clasificó una “virginica” como “versicolor” o viceversa).

El reporte muestra que, para “virginica”, el recall es 0.93, lo que significa que se clasificaron correctamente 14 de las 15.

La matriz de confusión confirma esto: un valor “1” en la celda (virginica → versicolor) indica ese único error.

 Paso 10: Visualizar resultados con gráfico de confusión (opcional)
Descripción:

Podemos dibujar la matriz de confusión para visualizar intuitivamente los aciertos y errores.

Usaremos seaborn.heatmap para darle formato.

````python
plt.figure(figsize=(5, 4))
sns.heatmap(
    cm_df,
    annot=True,
    cmap='Blues',
    fmt='d'
)
plt.title('Matriz de Confusión: Iris (Prueba)')
plt.xlabel('Predicción')
plt.ylabel('Etiqueta real')
plt.show()
````

La diagonal principal (15, 15, 14) en tonos más oscuros indica aciertos.

Solo aparece un valor fuera de la diagonal (1) donde “virginica” se clasificó como “versicolor”.

Conclusión del módulo práctico
A lo largo de este ejemplo con Iris hemos visto todo el flujo mínimo de ML supervisado:

Carga de datos reales desde scikit-learn → pd.DataFrame.

Exploración inicial (EDA) para entender si existen señales separables entre clases.

Selección de X (features) e y (etiqueta).

División en entrenamiento/prueba garantizando estratificación.

Escalado de variables numéricas (para mejorar convergencia y coherencia de pesos).

Elección e instanciación de un algoritmo supervisado (Regresión Logística).

Entrenamiento con fit.

Predicción con predict y, opcionalmente, predict_proba.

Evaluación con métricas (accuracy, precision, recall, F1) y matriz de confusión.

Visualización de resultados (heatmap).

Este mismo procedimiento se aplica a cualquier otro dataset de clasificación binaria o multiclase.

En problemas de regresión, cambiaríamos el modelo (por ejemplo, LinearRegression) y las métricas (por ejemplo, MSE, R²).
